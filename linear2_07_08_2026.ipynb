```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
x_obs = np.array([1, 2, 3, 4]).reshape(-1, 1)
y_obs = np.array([2, 2, 7, 8])

# Create regression model
model = LinearRegression()
model.fit(x_obs, y_obs)
a = round(model.coef_[0], 2)
b = round(model.intercept_)
y_pred = a*x_obs + b
mse = mean_squared_error(y_obs, y_pred)

print('Slope (שיפוע)', a)  # y = 2.3x - 1
print('Intercet (חיתוך)', b)
print('MSE', mse)

plt.axhline(0, color='black', linewidth=0.8, linestyle="--")  # x-axis
plt.axvline(0, color='black', linewidth=0.8, linestyle="--")  # y-axis
x = np.arange(0, 9)
plt.plot(x, a*x + b, label = f"y = {a}x + {b}")

plt.grid(color='gray', linestyle='--', linewidth=0.5)
plt.legend()

plt.scatter(x_obs, y_obs, color='red')
```
